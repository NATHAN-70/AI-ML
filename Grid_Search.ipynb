```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/drug200.csv')
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Cholesterol']=le.fit_transform(df['Cholesterol'])
df['BP']=le.fit_transform(df['BP'])
df['Drug']=le.fit_transform(df['Drug'])
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df[['Age','Na_to_K']]=sc.fit_transform(df[['Age','Na_to_K']])
```

```python
x=df.drop(["Drug"],axis=1)
y=df["Drug"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[None,10,20,40,50],
}
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=5,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,f1_score
f=f1_score(y_test,grid.predict(x_test), average='weighted')
f
```
