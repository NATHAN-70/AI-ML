```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/synthetic_premium_dataset.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for i in df.columns:
  if (df[i].dtype=='float64' or df[i].dtype=='int64') and i != 'purchased':
    df[i]=sc.fit_transform(df[[i]])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=le.fit_transform(df[[i]])
df
```

```python
x=df.drop('purchased',axis=1)
y=df['purchased']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Create separate encoders
le_country= LabelEncoder()
le_device_type= LabelEncoder()

# Fit encoders
le_country.fit(['TR', 'NL', 'DE', 'US'])
le_device_type.fit(['Desktop', 'Tablet', 'Mobile'])


def output(age, monthly_income, website_visits, time_spent_minutes, previous_purchases, is_student, country, device_type):

    # Transform inputs
    country_encoded = le_country.transform([country])[0]
    device_type_encoded = le_device_type.transform([device_type])[0]

    input_data = pd.DataFrame([[age, monthly_income, website_visits, time_spent_minutes, previous_purchases, is_student, country_encoded, device_type_encoded]],
                              columns=['age', 'monthly_income', 'website_visits', 'time_spent_minutes', 'previous_purchases', 'is_student', 'country', 'device_type'])

    prediction = rf.predict(input_data)[0]

    # Convert prediction to 'Yes' or 'No'
    if prediction == 1:
        return 'Yes he will purchase'
    else:
        return 'No he will not purchase'

print("Please enter the following details to predict whether a customer will purchase:")

try:
    age = float(input("Age (e.g., 30): "))
    monthly_income = float(input("Monthly Income (e.g., 2500): "))
    website_visits = float(input("Website Visits (e.g., 5): "))
    time_spent_minutes = float(input("Time Spent Minutes (e.g., 15.0): "))
    previous_purchases = float(input("Previous Purchases (e.g., 2): "))
    is_student = int(input("Is Student (0 for No, 1 for Yes): "))
    country = input("Country (TR, NL, DE, US): ").upper()
    device_type = input("Device Type (Desktop, Tablet, Mobile): ").capitalize()

    # Validate categorical inputs
    if country not in le_country.classes_:
        print(f"Country must be one of {list(le_country.classes_)}.")
    if device_type not in le_device_type.classes_:
        print(f"Device Type must be one of {list(le_device_type.classes_)}.")

    # Get prediction
    result = output(age, monthly_income, website_visits, time_spent_minutes, previous_purchases, is_student, country, device_type)

    print(f"\nBased on your inputs, the predicted purchase status is: {result}")


except ValueError as e:
    print(f"Invalid input: {e}. Please ensure all numerical inputs are valid numbers and categorical inputs are correct.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
