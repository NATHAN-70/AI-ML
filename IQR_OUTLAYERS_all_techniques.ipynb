```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Parental_Involvement']=le.fit_transform(df['Parental_Involvement'])
df['Access_to_Resources']=le.fit_transform(df['Access_to_Resources'])
df['Extracurricular_Activities']=le.fit_transform(df['Extracurricular_Activities'])
df['Motivation_Level']=le.fit_transform(df['Motivation_Level'])
df['Internet_Access']=le.fit_transform(df['Internet_Access'])
df['Family_Income']=le.fit_transform(df['Family_Income'])
df['Teacher_Quality']=le.fit_transform(df['Teacher_Quality'])
df['School_Type']=le.fit_transform(df['School_Type'])
df['Peer_Influence']=le.fit_transform(df['Peer_Influence'])
df['Learning_Disabilities']=le.fit_transform(df['Learning_Disabilities'])
df['Parental_Education_Level']=le.fit_transform(df['Parental_Education_Level'])
df['Distance_from_Home']=le.fit_transform(df['Distance_from_Home'])
df['Gender']=le.fit_transform(df['Gender'])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df['Hours_Studied']=scaler.fit_transform(df[['Hours_Studied']])
df['Exam_Score']=scaler.fit_transform(df[['Exam_Score']])
df['Attendance']=scaler.fit_transform(df[['Attendance']])
df['Parental_Involvement']=scaler.fit_transform(df[['Parental_Involvement']])
df['Access_to_Resources']=scaler.fit_transform(df[['Access_to_Resources']])
df['Sleep_Hours']=scaler.fit_transform(df[['Sleep_Hours']])
df['Previous_Scores']=scaler.fit_transform(df[['Previous_Scores']])
df['Motivation_Level']=scaler.fit_transform(df[['Motivation_Level']])
df['Family_Income']=scaler.fit_transform(df[['Family_Income']])
df['Teacher_Quality']=scaler.fit_transform(df[['Teacher_Quality']])
df['Physical_Activity']=scaler.fit_transform(df[['Physical_Activity']])

df
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=5)
df['Teacher_Quality']=imputer.fit_transform(df[['Teacher_Quality']])
df['Parental_Education_Level']=imputer.fit_transform(df[['Parental_Education_Level']])
df['Distance_from_Home']=imputer.fit_transform(df[['Distance_from_Home']])
df
```

```python
df.isnull().sum()
```

```python
x = df.drop('Exam_Score', axis=1)
y = df['Exam_Score']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression # Changed to LinearRegression
lr=LinearRegression() # Changed to LinearRegression
rfe=RFE(estimator=lr, n_features_to_select=20)
rfe.fit(x_train,y_train)
print('selected features:',rfe.support_)
print('feature ranking:',rfe.ranking_)
```

```python
selected_features=x_train.columns[rfe.support_]
print('selected features:')
for feature in selected_features:
  print(feature)
```

```python
df_selected=df[selected_features]
df_selected
```

```python
x=df_selected
y = df['Exam_Score']
```

```python
from scipy import LowLevelCallable
from scipy.stats import iqr
q1=df['Hours_Studied'].quantile(0.25)
q3=df['Hours_Studied'].quantile(0.75)
IQR=q3-q1
lower_bound=q1-1.5*IQR
upper_bound=q3+1.5*IQR
df['Hours_Studied']=np.where(df['Hours_Studied']<lower_bound,lower_bound,df['Hours_Studied'])
df['Hours_Studied']=np.where(df['Hours_Studied']>upper_bound,upper_bound,df['Hours_Studied'])
print(f"'Hours_Studied'column capped between {lower_bound} and {upper_bound},")
print("updated set")
print(df['Hours_Studied'].describe())
```
