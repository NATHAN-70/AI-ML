```python
import pandas as pd
df=pd.read_csv("/content/train.csv")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["trans_date_trans_time"]=le.fit_transform(df["trans_date_trans_time"])
df["merchant"]=le.fit_transform(df["merchant"])
df["category"]=le.fit_transform(df["category"])
df["first"]=le.fit_transform(df["first"])
df["last"]=le.fit_transform(df["last"])
df["gender"]=le.fit_transform(df["gender"])
df["city"]=le.fit_transform(df["city"])
df["state"]=le.fit_transform(df["state"])
df["job"]=le.fit_transform(df["job"])
df["dob"]=le.fit_transform(df["dob"])
df["trans_num"]=le.fit_transform(df["trans_num"])
df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df["cc_num"]=sc.fit_transform(df[["cc_num"]])
df["amt"]=sc.fit_transform(df[["amt"]])
df["zip"]=sc.fit_transform(df[["zip"]])
df["lat"]=sc.fit_transform(df[["lat"]])
df["long"]=sc.fit_transform(df[["long"]])
df["city_pop"]=sc.fit_transform(df[["city_pop"]])
df["unix_time"]=sc.fit_transform(df[["unix_time"]])
df["merch_lat"]=sc.fit_transform(df[["merch_lat"]])
df["merch_long"]=sc.fit_transform(df[["merch_long"]])
df["is_fraud"]=sc.fit_transform(df[["is_fraud"]])
df
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=5)
df['is_fraud']=knn.fit_transform(df[['is_fraud']])
df
```
