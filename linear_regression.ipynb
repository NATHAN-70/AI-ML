```python
import pandas as pd
df=pd.read_csv("/content/Salary (1).csv")
df.shape
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
x=df[["YearsExperience"]]
y=df[["Salary"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python

    years_experience = int(input("Enter the number of years of experience: "))
    predicted_salary = lr.predict([[years_experience]])
    print('predicted salary:',predicted_salary)
```

```python
import matplotlib.pyplot as plt
plt.scatter(years_experience,predicted_salary,color='red',label='Actual Data')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Years of Experience(linear regression)")

plt.show()
```
