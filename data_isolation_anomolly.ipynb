```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/personality_datasert (1).csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=le.fit_transform(df[i])
df[i]
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df['Time_spent_Alone']=sc.fit_transform(df[['Time_spent_Alone']])
df['Social_event_attendance']=sc.fit_transform(df[['Social_event_attendance']])
df['Going_outside']=sc.fit_transform(df[['Going_outside']])
df['Friends_circle_size']=sc.fit_transform(df[['Friends_circle_size']])
df['Post_frequency']=sc.fit_transform(df[['Post_frequency']])
df
```

```python
x=df.drop('Personality',axis=1)
```

```python
from sklearn.ensemble import IsolationForest
iso=IsolationForest()
iso.fit(x)
```

```python
y_pred=iso.predict(x)
y_pred=np.where(y_pred==-1,1,0)
df['isolation']=y_pred
df['isolation']
```

```python
print ("total anomalies detected:",df['isolation'].sum())
print(df[df['isolation']==1].head())
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y_pred)
plt.title("isolation")
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.show()
```
